```python
#Кривая Коха
import turtle
import time

turtle.speed('fastest')

def koh(size,n):
    if n == 0:
        turtle.forward(size)
    else:
        koh(size/3, n-1)
        turtle.left(60)
        koh(size/3, n-1)
        turtle.right(120)
        koh(size/3, n-1)
        turtle.left(60)
        koh(size/3, n-1)

koh(250, 3)
turtle.done()
```

```python
#Кривая Гильберта
turtle.speed('fastest')

def a(i, size):
    if i > 0:
        d(i - 1, size)
        turtle.goto(turtle.position()[0] + size, turtle.position()[1])
        a(i - 1, size)
        turtle.goto(turtle.position()[0], turtle.position()[1] + size)
        a(i - 1, size)
        turtle.goto(turtle.position()[0] - size, turtle.position()[1])
        c(i - 1, size)
        
def b(i, size):
    if i > 0:
        c(i - 1, size)
        turtle.goto(turtle.position()[0] - size, turtle.position()[1])
        b(i - 1, size)
        turtle.goto(turtle.position()[0], turtle.position()[1] - size)
        b(i - 1, size)
        turtle.goto(turtle.position()[0] + size, turtle.position()[1])
        d(i - 1, size)
        
def c(i, size):
    if i >0:
        b(i - 1, size)
        turtle.goto(turtle.position()[0], turtle.position()[1] - size)
        c(i - 1, size)
        turtle.goto(turtle.position()[0] - size, turtle.position()[1])
        c(i - 1, size)
        turtle.goto(turtle.position()[0], turtle.position()[1] + size)
        a(i - 1, size)
        
def d(i, size):
    if i > 0:
        a(i -1, size)
        turtle.goto(turtle.position()[0], turtle.position()[1] + size)
        d(i - 1, size)
        turtle.goto(turtle.position()[0] + size, turtle.position()[1])
        d(i - 1, size)
        turtle.goto(turtle.position()[0], turtle.position()[1] - size)
        b(i - 1, size)

a(3, 30)
turtle.done()
```

```python
#Кривая Серпинского
import math

turtle.speed('fastest')

def halfSierpinski(size, level):
    if level == 0:
        turtle.forward(size)
        return
    halfSierpinski(size, level - 1)
    turtle.left(45)
    turtle.forward(size * math.sqrt(2))
    turtle.left(45)
    halfSierpinski(size, level - 1)
    turtle.right(90)
    turtle.forward(size)
    turtle.right(90)
    halfSierpinski(size, level - 1)
    turtle.left(45)
    turtle.forward(size * math.sqrt(2))
    turtle.left(45)
    halfSierpinski(size, level - 1)

def sierpinski(size, level):
    halfSierpinski(size, level)
    turtle.right(90)
    turtle.forward(size)
    turtle.right(90)
    halfSierpinski(size, level)
    turtle.right(90)
    turtle.forward(size)
    turtle.right(90)

sierpinski(30,2)
turtle.done() 
```

```python
#Салфетка Серпинского
turtle.speed('fastest')

def drawTriangle(points):
    turtle.up()
    turtle.goto(points[0][0],points[0][1])
    turtle.down()
    turtle.goto(points[1][0],points[1][1])
    turtle.goto(points[2][0],points[2][1])
    turtle.goto(points[0][0],points[0][1])

def getMid(p1,p2):
    return ( (p1[0]+p2[0]) / 2, (p1[1] + p2[1]) / 2)

def sierpinski(points,level):
    drawTriangle(points)
    if level > 0:
        sierpinski([points[0],getMid(points[0], points[1]),getMid(points[0], points[2])],level-1)
        sierpinski([points[1],getMid(points[0], points[1]),getMid(points[1], points[2])],level-1)
        sierpinski([points[2],getMid(points[2], points[1]),getMid(points[0], points[2])],level-1)
 
myPoints = [[-200,-100],[0,100],[200,-100]]
sierpinski(myPoints,2)
turtle.done()
```

```python
#Ковер Серпинского
turtle.speed('fastest')

def sierpinski(n, m):
    if n == 0:
        turtle.color('black')
        turtle.begin_fill()
        for _ in range(4):
            turtle.forward(m)
            turtle.left(90)
        turtle.end_fill()
 
    else:
        for _ in range(4):
            sierpinski(n - 1, m / 3)
            turtle.forward(m / 3)
 
            sierpinski(n - 1, m / 3)
            turtle.forward(m / 3)
 
            turtle.forward(m / 3)
            turtle.left(90)
 
 
turtle.tracer(10)
sierpinski(2, 100)
turtle.done()
```

```python
#Дерево Пифагора
import time

turtle.speed("fastest")
turtle.left(90)

def tree(w, size):
    if w > 0:
        turtle.forward(size)
        turtle.right(45)
        tree(w - 1, size * (math.sqrt(2)/2))
        turtle.left(90)
        tree(w - 1, size * (math.sqrt(2)/2))
        turtle.right(45)
        turtle.forward(-(size))
        
tree(4, 70)
turtle.done()
```
